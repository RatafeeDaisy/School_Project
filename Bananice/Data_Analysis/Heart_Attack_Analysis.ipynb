```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
import plotly.express as px
from matplotlib import rcParams

plt.rcParams['font.family'] = 'SimHei'
plt.rcParams['axes.unicode_minus'] = False
sns.set(style='whitegrid', font='SimHei')
data = pd.read_csv('D:/School/School_Project/Bananice/Dataset/heart.csv')
```

```python
plt.figure(figsize=(20,10))
sns.countplot(x=data['age'])
plt.title('',fontsize=20)
plt.xlabel('',fontsize=20)
plt.ylabel('',fontsize=20)
plt.show()
```

```python
sns.displot(data['age'])
plt.title('',fontsize=20)
plt.xlabel('',fontsize=20)
plt.ylabel('',fontsize=20)
plt.show()
```

```python
s=data['sex'].value_counts().reset_index()
px.pie(s,names='index',values='sex',title='')
```

```python
c=data['cp'].value_counts().reset_index()
plt.figure(figsize=(10,5))
sns.barplot(x=c['index'],y=c['cp'])
plt.title('',fontsize=20)
plt.xlabel('',fontsize=20)
plt.ylabel('',fontsize=20)
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.histplot(data=data, x='age', hue='output')
plt.title("")
plt.xlabel('',fontsize=20)
plt.ylabel('',fontsize=20)
plt.show()
```
