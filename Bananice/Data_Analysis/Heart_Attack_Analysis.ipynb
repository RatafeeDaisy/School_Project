```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
import plotly.express as px
from matplotlib import rcParams

plt.rcParams['font.family'] = 'SimHei'
plt.rcParams['axes.unicode_minus'] = False
sns.set(style='whitegrid', font='SimHei')
data = pd.read_csv('D:/School/School_Project/Bananice/Dataset/heart.csv')
```

```python
data.sample(5)
```

```python
data.describe()
```

```python
data.duplicated().sum()
```

```python
plt.figure(figsize=(20,10))
sns.countplot(x=data['age'])
plt.title('年龄统计',fontsize=20)
plt.xlabel('岁数',fontsize=20)
plt.ylabel('数量',fontsize=20)
plt.show()
```

```python
sns.displot(data['age'])
plt.title('年龄分布',fontsize=20)
plt.xlabel('岁数',fontsize=20)
plt.ylabel('数量',fontsize=20)
plt.show()
```

```python
s=data['sex'].value_counts().reset_index()
px.pie(s,names='index',values='sex',title='男女比例饼图')
```

```python
c=data['cp'].value_counts().reset_index()
plt.figure(figsize=(10,5))
sns.barplot(x=c['index'],y=c['cp'])
plt.title('胸痛类型及数量',fontsize=20)
plt.xlabel('类型',fontsize=20)
plt.ylabel('数量',fontsize=20)
plt.show()
```

```python
sns.displot(data['trtbps'])
plt.title('血压分布情况',fontsize=20)
plt.xlabel('血压',fontsize=20)
plt.ylabel('数量',fontsize=20)
plt.show()
```

```python
sns.displot(data['chol'])
plt.title('胆固醇水平分布情况',fontsize=20)
plt.xlabel('胆固醇水平',fontsize=20)
plt.ylabel('数量',fontsize=20)
plt.show()
```

```python
sns.displot(data['thalachh'])
plt.title('心率分布情况',fontsize=20)
plt.xlabel('心率',fontsize=20)
plt.ylabel('数量',fontsize=20)
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.histplot(data=data, x='age', hue='output')
plt.title("年龄是否影响心脏病发作")
plt.xlabel('年龄',fontsize=20)
plt.ylabel('数量',fontsize=20)
plt.show()
```

```python
v=pd.crosstab(data['age'],data['output']).reset_index()
v.columns=['年龄','低风险','高风险']
px.line(v,v['年龄'],v['高风险'],title='心脏病高风险是否与年龄有关')
```

```python
px.line(v,v['年龄'],v['低风险'],title='心脏病低风险是否与年龄有关')
```

```python
plt.figure(figsize=(15,5))
sns.lineplot(y='trtbps',x='age',data=data)
plt.title('血压随年龄的变化',fontsize=20)
plt.xlabel('年龄',fontsize=20)
plt.ylabel('血压',fontsize=20)
plt.show()
```

```python
plt.figure(figsize=(15,5))
sns.lineplot(y='chol',x='age',data=data)
plt.title('胆固醇水平随年龄的变化',fontsize=20)
plt.xlabel('年龄',fontsize=20)
plt.ylabel('胆固醇水平',fontsize=20)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.lineplot(x='age',y='chol',hue='output',data=data)
plt.title('年龄和胆固醇与心脏病发作的关系',fontsize=20)
plt.xlabel('年龄',fontsize=20)
plt.ylabel('胆固醇水平',fontsize=20)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.lineplot(x='age',y='trtbps',hue='output',data=data)
plt.title("年龄和血压与心脏病发作的关系",fontsize=20)
plt.xlabel('年龄',fontsize=20)
plt.ylabel('血压',fontsize=20)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.lineplot(x="age",y="thalachh",hue="output",data=data)
plt.title("年龄和最大心率高低与心脏病发作的关系",fontsize=20)
plt.xlabel('年龄',fontsize=20)
plt.ylabel('心率',fontsize=20)
plt.show()
```
