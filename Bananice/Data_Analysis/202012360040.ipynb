```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
rc = {'font.sans-serif': 'SimHei'}
sns.set(style='whitegrid', rc=rc)
```

```python
data = pd.read_csv('D:/School/Bananice/数据集/sh.csv')
data.head(5)
data.info()
```

```python
data.head(5)
```

```python
data.info()
```

```python
plt.figure(dpi=100,figsize=(15,8))
plt.title('上海二手房各区域总体均价',loc='center',pad=15,fontsize=20)
sns.barplot(x=data['区域'],y=data['价格（W）'],data=data)
```

```python
plt.figure(dpi=100,figsize=(15,8))
plt.title('上海二手房各区域平均单价',loc='center',pad=15,fontsize=20)
sns.barplot(x=data['区域'],y=data['单价（平方米）'],data=data)
```

```python
sns.relplot(data=data,x='单价（平方米）',y='价格（W）',col='区域',col_wrap=4)
```

```python
sns.relplot(x="朝向", y="面积", data=data,hue="区域",sizes=(20, 200),aspect=4,height=5)
```

```python
sns.relplot(x="面积", y="价格（W）",hue='区域',data=data,aspect=2,height=4)
```

```python
a = data.corr()
sns.heatmap(a,annot=True,fmt='.2f',cmap='Purples')
```
